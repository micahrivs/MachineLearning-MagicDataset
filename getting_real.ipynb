```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf

# Import our input dataset
attrition_df = pd.read_csv('https://static.bc-edx.com/data/dl-1-2/m21/lessons/2/HR-Employee-Attrition.csv')
attrition_df
```

```python
# Generate our categorical variable lists
df_cat= attrition_df.dtypes[attrition_df.dtypes=="object"].index.tolist()
```

```python
# Check the number of unique values in each column
df_cat
```

```python
attrition_df[df_cat].nunique()
```

```python
# Create a OneHotEncoder instance
enc=OneHotEncoder(sparse=False)
# Fit and transform the OneHotEncoder using the categorical variable list
encode_df=pd.DataFrame(enc.fit_transform(attrition_df[df_cat]))

# Add the encoded variable names to the dataframe
encode_df.columns=enc.get_feature_names_out(df_cat)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
attritiond_df=attrition_df.merge(encode_df, left_index=True, right_index=True)
#option2
# attrition_df=pd.concat([attrition_df,encode_df],axis=1)
attrition_df=attrition_df.drop(df_cat,axis=1)
attrition_df.head()
```

```python
attrition_df=attrition_df.drop(columns=["Unnamed: 0"])
```

```python
# Split our preprocessed data into our features and target arrays
y= encode_df['Attrition_Yes'].values
X=encode_df.drop(columns=['Attrition_Yes','Attrition_No']).values
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train,y_test=train_test_split(X,y,random_state=42)
```

```python
# Create a StandardScaler instances
scaler=StandardScaler()

# Fit the StandardScaler
X_scaler=scaler.fit(X_train)

# Scale the data
X_train_scaled=X_scaler.transform(X_train)
X_test_scaled=X_scaler.transform(X_test)
```

```python
len(X_train_scaled[0])
```

```python
# Define the model - deep neural net
num_of_features=len(X_train_scaled[0])

nn_model=tf.keras.models.Sequential()
# First hidden layer
nn_model.add(
    tf.keras.layers.Dense(units=40, activation='relu',input_dim=num_of_features)
)

# Second hidden layer
nn_model.add(
    tf.keras.layers.Dense(units=10, activation='relu')
)

# Output layer

nn_model.add(
    tf.keras.layers.Dense(units=1, activation='sigmoid')
)


# Check the structure of the model
```

```python
# Compile the model
nn_model.compile(loss="binary_crossentropy", optimizer="adam",  metrics=["accuracy"])
```

```python
# Train the model

fit_model=nn_model.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_acc=nn_model.evaluate(X_test_scaled,y_test,verbose=2)
```
